### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
import pandas as pd

athletic_sales_2020 = pd.read_csv('athletic_sales_2020.csv')
athletic_sales_2021 = pd.read_csv('athletic_sales_2021.csv')

```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

combined_sales = pd.concat([athletic_sales_2020, athletic_sales_2021], axis = 'rows', join = 'inner').reset_index(drop=True)

combined_sales.head()



```

```python
# Check if any values are null.
#combined_sales.isnull().any()
combined_sales.isna().sum()
```

```python
# Check the data type of each column
combined_sales.info()
```

```python
# Convert the "invoice_date" to a datetime datatype

# Convert the "invoice_date" column to datetime data type
combined_sales['invoice_date'] = pd.to_datetime(combined_sales['invoice_date'], errors ='coerce')
combined_sales
```

```python
# Confirm that the "invoice_date" data type has been changed.

combined_sales.info()
combined_sales
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
#Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold = combined_sales.groupby(['region','state','city'])\
            ['units_sold'].agg(Total_Products_Sold=('sum'))



# Show the top 5 results.
total_products_sold.sort_values(by=(['Total_Products_Sold']), ascending=False).head()
```

#### Using `pivot_table`

```python

# Show the number products sold for region, state, and city.
total_products_sold_region = pd.pivot_table(combined_sales, index=['region','state','city'],
                                     values='units_sold',
                                     aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
total_prdcts = total_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
total_prdcts.sort_values(by=("Total_Products_Sold"), ascending=False).head()

```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.



filtered_womens_footwear = combined_sales[combined_sales["product"].str.contains("Women's Athletic Footwear")]
filtered_womens_footwear
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_footwear_products_sold = filtered_womens_footwear.groupby(['retailer','region','state','city'])\
            ['units_sold'].agg(units_sold=('sum'))


# Show the top 5 results.
women_footwear_products_sold.sort_values(by=(['units_sold']), ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

women_footwear_products_sold_retailer = pd.pivot_table(filtered_womens_footwear, index=['retailer','region','state','city'],
                                     values='units_sold',
                                     aggfunc='sum')

women_footwear_products_sold_retailer
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

total_women_footwear = women_footwear_products_sold_retailer.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

total_women_footwear

# Show the top 5 results.
total_women_footwear.sort_values(by=("Womens_Footwear_Units_Sold"), ascending=False).head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

womens_footwear_cnt = pd.pivot_table(filtered_womens_footwear, 
                                     index=['invoice_date'],
                                     values='total_sales',
                                     aggfunc='sum')
womens_footwear_cnt
```

```python
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sold = womens_footwear_cnt.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})



womens_footwear_sold.head()
```

```python
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

womens_footwear_sold = womens_footwear_sold.sort_values(by='total_sales', ascending=False)

# Show the top 5 results.

womens_footwear_sold.head()

# Optional: Rename the "total_sales" column to "Total Sales"
#most_women_products_sold = women_footwear_products_sold_retailer.rename(columns={'total_sales': 'TotalSales'})
womens_footwear_sold = womens_footwear_cnt.rename(columns={"total_sales": "TotalSales"})
womens_footwear_sold

# Show the table.
#most_women_products_sold.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

# Assuming df is your DataFrame containing the pivot table
# Resample to daily bins and sum the sales for each day
daily_sales = womens_footwear_sold.resample('D').sum()



# Sort the resampled pivot table in ascending order on "Total Sales".

#womens_footwear_sold = womens_footwear_cnt.rename(columns={"total_sales": "TotalSales"})

daily_sales = daily_sales.sort_values(by='TotalSales', ascending=False)


daily_sales
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales_sales = womens_footwear_sold.resample('W').sum()


# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_sales = daily_sales.sort_values(by='TotalSales', ascending=False)

weekly_sales.head()
```
